# Test for Airflow API

```python
import requests
import base64

import pandas as pd
```

## Encode default password

```python
original_pass = "airflow:airflow"
encoded_token = base64.b64encode(original_pass.encode()).decode()
```

```python
API_URL = "http://localhost:8080/api/v1"
HEADER= {
    'Authorization': f'Basic {encoded_token}',
}
```

## List DAG runs

```python
resp = requests.get(
    url=f"{API_URL}/dags",
    headers=HEADER,
    params={
        'limit': 5,
    }
)
print(resp.status_code)
```

```python
pd.DataFrame(resp.json()['dags'])
```

```python
## Dag Info
```

```python
def get_dag_runs_from_dagid(dag_id: str):
    resp = requests.get(
        url=f"{API_URL}/dags/{dag_id}/dagRuns",
        headers=HEADER,
        params={}
    )
    return resp
```

```python
resp = get_dag_runs_from_dagid('play_woodwings')
```

```python
pd.DataFrame(resp.json()['dag_runs'])
```

## Trigger a new run

```python
def new_dag_run_for_dagid(dag_id: str):
    resp = requests.post(
        url=f"{API_URL}/dags/{dag_id}/dagRuns",
        headers=HEADER,
        json={
            'conf': {}, # conf for the dag run
        }
    )
    return resp
```

```python
get = new_dag_run_for_dagid('play_woodwings')
```
